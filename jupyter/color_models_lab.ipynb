# Image Processing Lab 2 (Color Model Conversion)

Importing Libraries

```python
import cv2
import matplotlib.pyplot as plt
```

Importing an image

```python
img = cv2.imread('../images/image3.JPG')
```

Converting the image into different color models

```python
rgb_image = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
hsv_image = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

Displalying the different images

```python
# Base Image
# cv2.imshow('Base Image', img)
plt.imshow(img)
plt.title('Base Image')
plt.show()
```

```python
# RGB Image
plt.imshow(rgb_image)
plt.title('RGB Image')
plt.show()
```

```python
# HSV Image
plt.imshow(hsv_image)
plt.title('HSV Image')
plt.show()
```

```python
# Grayscale Image
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.show()
```

Creating image histograms

```python
# For base image
plt.title('Base Image')
plt.xlabel('Intensity')
plt.ylabel('Number of Pixels')
colors = ('b', 'g', 'r')
for i, col in enumerate(colors):
    hist = cv2.calcHist([img], [i], None, [256], [0, 256])
    plt.plot(hist, color = col)
plt.show()
```

```python
# For RGB image
plt.title('RGB Image')
plt.xlabel('Intensity')
plt.ylabel('Number of Pixels')
colors = ('r', 'g', 'b')
for i, col in enumerate(colors):
    hist = cv2.calcHist([rgb_image], [i], None, [256], [0, 256])
    plt.plot(hist, color = col)
plt.show()
```

```python
# For HSV image
plt.title('HSV Image')
plt.xlabel('Intensity')
plt.ylabel('Number of Pixels')
channels = ('Hue', 'Saturation', 'Value')
colors = ('r', 'g', 'b')
for i, (chan, col) in enumerate(zip(channels, colors)):
    hist = cv2.calcHist([hsv_image], [i], None, [256], [0, 256])
    plt.plot(hist, color = col, label=chan)
plt.show()
```

```python
# For grayscale image
plt.title('Grayscale Image')
plt.xlabel('Intensity')
plt.ylabel('Number of Pixels')
hist = cv2.calcHist([gray_image], [0], None, [256], [0, 256])
plt.plot(hist, color = 'k')
plt.show()
```
