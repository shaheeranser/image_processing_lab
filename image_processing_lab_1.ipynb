# Image Processing Lab 1 (Introduction)

Importing Libraries

```python
import cv2
import matplotlib.pyplot as plt
```

Importing An image

```python
img = cv2.imread('images/image1.JPG') # The folder isn't pushed to the repo
```

Displaying The Image

```python
# cv2.imshow('OpenCV Image', img)
# (not working on jupyter notebooks)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Image using Matplotlib')
plt.show()
```

Displaying The Image In Grayscale

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2GRAY), cmap='gray')
plt.title('Grayscale Image using Matplotlib')
plt.show()
```
